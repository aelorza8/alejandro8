## ¿Is there a strong relationship between fixed acidity and citric acid?

## Conclusion

## Through the use of linear regression, we sought to try to analyze some variables from another, so the following was obtained:

```python
for i in df['quality'].index:
    if df['quality'][i]<5:
        df['quality'][i] = '<5'
    elif df['quality'][i]>=5 and df['quality'][i]<=6:
        df['quality'][i] ='5-6'
    else:
        df['quality'][i]='>6'                  
```

```python
df 
```

```python
counts = df['quality'].value_counts()
counts = counts.to_frame()
type(counts)
```

```python
counts['%'] = (counts.iloc[:,:]/counts.loc[:,'quality'].sum(axis=0))*100
counts
```

```python
counts = counts.reindex(['5-6','>6','<5',])
counts = counts.reset_index()
counts
```

```python
counts = counts.rename({'index':'Group quality','quality':'quality'},
                       axis = 1)
counts
```

## ¿Which porcentage of subgroups of qualities are more present in this data base?

```python
plt.figure(figsize=(14,10))
plt.pie(counts['quality'], labels = counts['Group quality'], autopct = '%1.1f%%')
plt.title('Quality groups')
plt.legend()
```

## Conclusion

## By using a pie chart we can classify the different qualities found in the database. Also using pivot tables and knowing that there are 1113 data, the classification is as follows:
## Wines with quality less than 5 represent 3.41%, wines with quality or 6 represent 82.67% and those with quality greater than 6 represent 13.9%.

## ¿Can the density of the wine be calculated using the fixed acidity as a predictor variable?

## Density = 0.99049119 + 0.00075071 fixed acidity

```python
plt.figure(figsize=(10,6))

ax= sns.regplot(x='density',
               y='fixed acidity',
               data=df,
               color = 'green',
                marker= '+',
              )
ax.set(xlabel = 'density',
       ylabel = 'fixed acidity')
#ax.text(10,200,'density = 0.99049119 + 0.00075071 fixed acidity',
        #horizontalalignment = 'center', size= 'small',
        #color= 'black', weight= 'semibold')

ax.set_title('Regression plot of density vs fixed acidity')
plt.show()
```
