```python
df['quality'].describe()
```

```python
for i in df['quality'].index:
    if df['quality'][i]<5:
        df['quality'][i] = '<5'
    elif df['quality'][i]>=5 and df['quality'][i]<=6:
        df['quality'][i] ='5-6'
    else:
        df['quality'][i]='>6'                  
```

```python
df 
```

```python
counts = df['quality'].value_counts()
counts = counts.to_frame()
type(counts)
```

```python
counts['%'] = (counts.iloc[:,:]/counts.loc[:,'quality'].sum(axis=0))*100
counts
```

```python
counts = counts.reindex(['5-6','>6','<5',])
counts = counts.reset_index()
counts
```

```python
counts = counts.rename({'index':'Group quality','quality':'quality'},
                       axis = 1)
counts
```

## ¿Which porcentage of subgroups of qualities are more present in this data base?

```python
plt.figure(figsize=(14,10))
plt.pie(counts['quality'], labels = counts['Group quality'], autopct = '%1.1f%%')
plt.title('Quality groups')
plt.legend()
```

```python
from sklearn import linear_model
```

```python
y= pd.DataFrame(df['density'])
x =pd.DataFrame(df['fixed acidity'])
lm= linear_model.LinearRegression()
model_lm=lm.fit(x,y)
model_lm
```

```python
print('The y intercetpt, B0 is =' , model_lm.intercept_)
print('the coef., b1 is =' , model_lm.coef_)
print('theR^2, is =', model_lm.score(x,y))
```
