##  ¿Is it possible to predict the fixed acidity with the variable citric acid?

```python
sns.lmplot(x = 'citric acid',
           y = 'fixed acidity',
           data=df,
           fit_reg = True,
          )
plt.title('Relation between Fixed Acidity and Citric acid', fontsize = 20)
           
```

## Conclusion

##  ¿There is a level of sulphate that has more frequency in the quality of the wine??

```python
sns.swarmplot (x ='quality',
                y= 'sulphates',
                data = df
               )
plt.title ('Swarm plot de Alcohol vs Quality ', fontsize =20)
```
