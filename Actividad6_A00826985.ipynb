##  6 ¿Is it possible to predict the fixed acidity with the variable citric acid?

```python
sns.lmplot(x = 'citric acid',
           y = 'fixed acidity',
           data=df,
           fit_reg = True,
          )
plt.title('Relation between Fixed Acidity and Citric acid', fontsize = 20)
           
```
