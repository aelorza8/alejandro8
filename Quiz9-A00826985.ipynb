```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
%ls datasets/
```

```python
df = pd.read_csv('datasets/antropometria-dataset(9).csv')
df.head()
```

```python
df.dropna(inplace= True)
df
```

```python
y= pd.DataFrame(df['cadera'])
x =pd.DataFrame(df['peso'])
lm= linear_model.LinearRegression()
model_lm=lm.fit(x,y)
model_lm
```

```python
print('The y intercetpt, B0 is =' , model_lm.intercept_)
print('the coef., b1 is =' , model_lm.coef_)
print('theR^2, is =', model_lm.score(x,y))
```
